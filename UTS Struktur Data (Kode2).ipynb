```python
# ============================================================
#  SISTEM UNDO / REDO TEXT EDITOR
#  Implementasi Stack menggunakan Linked List
#  Mata Kuliah: Struktur Data - UTS Project
# ============================================================

# ─── Node Linked List ────────────────────────────────────────
class Node:
    def __init__(self, data):
        self.data = data   # isi teks (snapshot)
        self.next = None

# ─── Stack ───────────────────────────────────────────────────
class Stack:
    def __init__(self, nama: str):
        self.top  = None   # pointer ke elemen paling atas
        self.size = 0
        self.nama = nama   # label stack (Undo / Redo)

    def is_empty(self) -> bool:
        return self.top is None

    # Push: tambah elemen ke atas stack
    def push(self, data: str):
        node      = Node(data)
        node.next = self.top
        self.top  = node
        self.size += 1

    # Pop: ambil dan hapus elemen teratas
    def pop(self) -> str:
        if self.is_empty():
            return None
        data      = self.top.data
        self.top  = self.top.next
        self.size -= 1
        return data

    # Peek: lihat elemen teratas tanpa menghapus
    def peek(self) -> str:
        if self.is_empty():
            return None
        return self.top.data

    # Display: tampilkan seluruh isi stack
    def display(self):
        if self.is_empty():
            print(f"   (Stack {self.nama} kosong)")
            return
        curr  = self.top
        items = []
        while curr:
            items.append(curr.data)
            curr = curr.next
        for i, item in enumerate(items):
            label = " ← TOP" if i == 0 else ""
            preview = item if len(item) <= 40 else item[:37] + "..."
            print(f"   [{i+1}] \"{preview}\"{label}")


# ─── Text Editor ─────────────────────────────────────────────
class TextEditor:
    def __init__(self):
        self.teks_sekarang = ""   # kondisi teks saat ini
        self.undo_stack    = Stack("Undo")
        self.redo_stack    = Stack("Redo")

    # Tulis / edit teks (aksi baru)
    def tulis(self, teks_baru: str):
        # Simpan kondisi sebelumnya ke undo stack
        self.undo_stack.push(self.teks_sekarang)
        # Setiap aksi baru menghapus redo stack (LIFO semantics)
        self.redo_stack = Stack("Redo")
        self.teks_sekarang = teks_baru
        print(f"\n[✏]  Teks diperbarui.")
        print(f"     Isi sekarang : \"{self.teks_sekarang}\"")

    # Undo: kembalikan ke kondisi sebelumnya
    def undo(self):
        if self.undo_stack.is_empty():
            print("\n[!]  Tidak ada aksi yang dapat di-Undo.")
            return
        # Simpan kondisi sekarang ke redo stack sebelum undo
        self.redo_stack.push(self.teks_sekarang)
        self.teks_sekarang = self.undo_stack.pop()
        print(f"\n[↩]  Undo berhasil.")
        print(f"     Isi sekarang : \"{self.teks_sekarang}\"")

    # Redo: ulangi aksi yang sudah di-undo
    def redo(self):
        if self.redo_stack.is_empty():
            print("\n[!]  Tidak ada aksi yang dapat di-Redo.")
            return
        # Simpan kondisi sekarang ke undo stack sebelum redo
        self.undo_stack.push(self.teks_sekarang)
        self.teks_sekarang = self.redo_stack.pop()
        print(f"\n[↪]  Redo berhasil.")
        print(f"     Isi sekarang : \"{self.teks_sekarang}\"")

    # Peek: lihat elemen teratas setiap stack
    def peek_stacks(self):
        print(f"\n[i]  Teks sekarang  : \"{self.teks_sekarang}\"")
        undo_top = self.undo_stack.peek()
        redo_top = self.redo_stack.peek()
        print(f"     Undo (TOP)    : \"{undo_top}\"" if undo_top is not None else "     Undo Stack   : (kosong)")
        print(f"     Redo (TOP)    : \"{redo_top}\"" if redo_top is not None else "     Redo Stack   : (kosong)")

    # Display: tampilkan riwayat lengkap
    def display_history(self):
        print("\n" + "═" * 52)
        print("  RIWAYAT EDITOR")
        print("═" * 52)
        print(f"  Teks Aktif : \"{self.teks_sekarang}\"")
        print("─" * 52)
        print("  UNDO STACK (riwayat ke belakang):")
        self.undo_stack.display()
        print("─" * 52)
        print("  REDO STACK (riwayat ke depan):")
        self.redo_stack.display()
        print("═" * 52)
```
